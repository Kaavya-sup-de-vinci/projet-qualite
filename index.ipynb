```python
#Nous continuons avec un deuxième dataset du lien du PDF également, ou nous allons retrouver les pollutions par jour de l'année 2023
#Ce dataset est très interessant étant donné qu'il nous montre 5 pollutions différentes à chaque jour de l'année 

fact = pd.read_csv("qualite_air.csv")

#Nous supprimons ces 3 colonnes étant donné que nous créons la 2ème table de dimensions, nous reviendrons sur les colonnes supprimées par la suite
fact = fact.drop(columns=["valeur","Latitude","Longitude"],inplace=True)
fact
```

```python
#Nous avons crée ici une clé primaire qui va nous faciliter les relations 
fact["ID_qualite"] = fact["Date de début"] + fact["Polluant"] + fact["Code postal"]
#fact.to_csv("dim_qualite.csv")
```

```python
#Nous avons ici 2 sources de données externes qui va nous sortir un output table de dimension avec le nb de lignes de transport dans ces 2 villes

data_lorient = pd.read_csv("routes_Lorient.csv",delimiter=";")
data_lorient.drop(columns="route_desc",inplace=True)
```

```python
#Colonne qui ne va aps être utile dans notre cas de figure
data_lorient.drop(columns="route_long_name",inplace=True)
data_lorient
```

```python
data_rennes = pd.read_csv("métro et bus Rennes.csv",delimiter=";")
data_rennes
```

```python
#Nous rajoutons le code postal dans cette table de dimension, afin d'appeller depuis la table des faits cette table de dimension
#Par le code postal qui nous indiquera automatiquement le nombre de transport dans nos analyses

lorient_lines = data_lorient.value_counts("route_id").sum()
rennes_lines = data_rennes.value_counts().sum()
data = {
    "Ville": ["Lorient", "Rennes"],
    "Nombre de lignes de transport": [lorient_lines, rennes_lines],
     "Code postal": [56100, 35000]
}
df = pd.DataFrame(data)

df

#df.to_csv("dim_nb_transport.csv")
```
